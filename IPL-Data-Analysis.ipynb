```python
def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]
```

```python
def get_toss_match_count(team):
    mask = (data['toss_winner'] == data['winner']) & (data['toss_decision'] == 'bat') & (data['toss_winner'] == team)
    return data[mask].shape[0]
```
